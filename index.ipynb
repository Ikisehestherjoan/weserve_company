DATA TRANSFORMATION AND CLEANING

```python
# CONVERTING THE DATATYPE FOR THE CALLID FROM OBJECT TO INT TO ENABLE MERGING OF THE TWO DATA SET
call_details_copy['callID']= call_details_copy['callID'].astype(str).astype(int)

```

```python
call_details_copy.info()
```

CREATE A NEW DATA BY MERGING THE TWO EXSITING DATA 

```python
# MERGING THE TWO DATASET BY A COMMON COLUMN WHCH IS THE ID ON THE CALL_LOG_COPY AND CALLID ON CALLDETAILS_COPY
merged_details=pd.merge(call_log_copy,call_details_copy, left_on='id', right_on='callID')
merged_details=merged_details.set_index('id')
```

```python
merged_details
```

```python
# Filling the missing values in the assignedTo with their coressponding agentID
merged_details['assignedTo'].fillna(merged_details['agentID'], inplace=True)
```

```python
# droping the callID since its the same with the id
merged_details.drop(columns=['callID'], inplace=True)
```

```python
merged_details.head()
```

```python

# replacing the NaN in resolutionDurationInHours to 0 since the status isnt closed
merged_details['resolutionDurationInHours']=merged_details['resolutionDurationInHours'].fillna(0)
merged_details
```

```python
# replacing the In-bound to Inbound in the callType column
# merged_details['callType']=merged_details['callType'].replace(['In-bound','Inbound'])
# s.replace({'a': None}
merged_details['callType']=merged_details['callType'].replace({'in-bound':'Inbound'})
```

```python
# convertiing the float type of resolutionDurationInHours to int
merged_details['resolutionDurationInHours'] = merged_details['resolutionDurationInHours'].astype(int)
```

```python
# Replacing the cases of the status to be uniform
merged_details['status']=merged_details['status'].replace(['CLOSED','pEnding'],['closed','pending'])
```

```python
#     -- -1. Total Number of Calls Resolved vs Total Number of Calls Received

query='''
       
        SELECT
            SUM(CASE WHEN status = 'resolved' THEN 1 ELSE 0 END) AS TotalResolvedCalls,
            COUNT(*) AS TotalCallsReceived
        FROM
            call_details;

'''
```

```python
cursor.execute(query)
query_1=cursor.fetchall()
```

```python
cursor.description
```

```python
calls_vs_received =pd.DataFrame(query_1,columns=['totalresolvedcalls','totalcallsreceived'])
calls_vs_received
```

```python
# -- -------2. Total Number of calls received vs Total Number of calls assigned/resolved
query_2='''
        SELECT
            COUNT(*) AS TotalCallsReceived,
            SUM(CASE WHEN status IN ('closed','pending', 'Resolved') THEN 1 ELSE 0 END) AS TotalAssignedOrResolvedCalls
        FROM
            call_details;

'''

```

```python
cursor.execute(query_2)
total_calls_assignedorresolved = cursor.fetchall()
cursor.description
```

```python
total_calls_assignedorresolved=pd.DataFrame(total_calls_assignedorresolved,columns=['totalcallsreceived','totalassignedorresolvedcalls'])
total_calls_assignedorresolved
```

```python
# -- -3. The total and average call duration for each agent, and the grade level of these agents.
query_3 = '''
                SELECT 
                        "agentsGradeLevel",
                        ROUND(SUM("callDurationInSeconds"),2) AS TotalCallduration, 
                        ROUND(AVG("callDurationInSeconds"),2) AS AvgCallDuration
                FROM call_details
                GROUP BY "agentsGradeLevel"
                ORDER BY "agentsGradeLevel"

        '''


```

```python
cursor.execute(query_3)
tot_avg_of_agents= cursor.fetchall()
cursor.description
```

```python
tot_avg_of_agents =pd.DataFrame(tot_avg_of_agents,columns=['agentsGradeLevel','totalcallduration','avgcallduration'])
tot_avg_of_agents
```

```python
#  -- --4.The earliest and latest closed and resolved calls, and the grade levels of the agents who resolved these cases.
query_4 =    '''
            SELECT
                    MIN("resolutionDurationInHours") AS EarliestClosedCall,
                    MAX("resolutionDurationInHours") AS LatestClosedCall,
                    "agentsGradeLevel" AS AgentGradeLevel
        FROM
            call_details
        WHERE
                    "status" IN ('closed', 'resolved')
        GROUP BY
                    "agentsGradeLevel"
        ORDER BY "agentsGradeLevel" ;

'''

```

```python
cursor.execute(query_4)
ear_last_closed_res= cursor.fetchall()
cursor.description 
```

```python
ear_last_closed_res=pd.DataFrame(ear_last_closed_res,columns=['earliestclosedcall','latestclosedcall','agentgradelevel'])
ear_last_closed_res
```

```python
# Total gradelevel by agents
query_5 ='''
            SELECT 
                    "agentsGradeLevel",
                    COUNT("agentsGradeLevel") AS Total_grade_lev
            FROM call_details
            GROUP BY "agentsGradeLevel"
            ORDER BY "agentsGradeLevel";
'''
```

```python
cursor.execute(query_5)
tot_gradelevel_by_agent = cursor.fetchall()
cursor.description
```

```python
tot_gradelevel_by_agent =pd.DataFrame(tot_gradelevel_by_agent,columns=['agentsGradeLevel','total_grade_lev'])
tot_gradelevel_by_agent
```

```python
# -- 6.-- THE TOTAL NUMBER OF THE CALL TYPE 
query_6 = '''
                SELECT "callType",COUNT("callType") AS cnt_calltype
                FROM call_details
                GROUP BY "callType";

            '''
cursor.execute(query_6)
tot_calltype=cursor.fetchall()
tot_calltype = pd.DataFrame(tot_calltype,columns=['calltype','cnt_calltype'])
tot_calltype
```

```python
# -- 7.-- -- HOW MANY ASSIGNED JOBS WERE GIVEN TO THE DIFFERENT AGENTS
 
query_7 = '''
            SELECT "assignedTo", COUNT("assignedTo") AS assigned_count FROM call_details
            GROUP BY "assignedTo"
            ORDER BY  assigned_count DESC

            '''
cursor.execute(query_7)
tot_assignedjob=cursor.fetchall()
tot_assignedjob = pd.DataFrame(tot_assignedjob,columns=['assignedTo','assigned_count'])
tot_assignedjob
```
